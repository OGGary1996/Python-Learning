# Algorithm Analysis
Analyze the efficiency of algorithms

1. Big O notation is a mathematical notation used to describe the worst-case time
complexity of an algorithm. It tells us how the runtime or space requirements grow as the input size increases.

2. Big O helps compare the efficiency of algorithms, especially for large inputs. A more efficient algorithm can handle larger datasets much faster.

For example:

* A O(nÂ²) algorithm on an input of 1,000,000 may take years.
* A O(log n) algorithm can process the same input in milliseconds.

# Properties of Big -O
1. if f(n) is a sum of serveral terms, the one with the largest growth rate is kept, and all others omitted.
2. if f(n) is a product of serveral factors, any constants are omitted - which means you can drop coefficients.

```python
def constant_time(n):
    """O(1) - Constant Time"""
    return n * 2  # Always takes the same time regardless of input size

def linear_time(arr):
    """O(n) - Linear Time"""
    for num in arr:
        print(num)  # Time grows linearly with input size

def quadratic_time(arr):
    """O(n^2) - Quadratic Time"""
    for i in arr:
        for j in arr:
            print(i, j)  # Time grows quadratically

def logarithmic_time(n):
    """O(log n) - Logarithmic Time"""
    while n > 1:
        print(n)
        n //= 2  # Reduces problem size in half each step

def linear_log_time(n):
    """O(n log n) - Linearithmic Time"""
    for i in range(n):
        j = 1
        while j < n:
            print(i, j)
            j *= 2  # Nested loop with logarithmic inner loop

# Test functions
print("O(1) Example:")
constant_time(10)

print("\nO(n) Example:")
linear_time([1, 2, 3, 4])

print("\nO(n^2) Example:")
quadratic_time([1, 2, 3])

print("\nO(log n) Example:")
logarithmic_time(16)

print("\nO(n log n) Example:")
linear_log_time(8)
```
